# KNN 구현

```python
from random import randint
N = 100
D = list((randint(1, 100), randint(1, 100)) for _ in range(N))
C = list(0 if sum(d) > 100 else 1 for d in D)
sample = (50, 50)
```

```python
import matplotlib.pyplot as plt
from math import sqrt

X = [d[0] for d in D]
Y = [d[1] for d in D]
similarity = [0.0 for _ in range(N)]
colorMap = ["r", "g", "b"]

# class o => r, class 1 => g
for i in range(len(C)):
    plt.scatter(X[i], Y[i],
                    facecolor = "none",
                    edgecolors = colorMap[C[i]],
                    linewidths = .5)
    similarity[i] = cos(D[i], sample)
plt.scatter(sample[0], sample[1],
           facecolor = "none",
           edgecolors = colorMap[-1])
plt.show()
```

### 유사도 계산 두가지 방법

```python
# rel(d, q) => sim(d,q), in VS
def euc(x1, x2):
    # linalg.norm
    return sqrt((x1[0] - x2[0])**2 + (x1[1] - x2[1])**2)

def cos(x1, x2):
    # np.dot
    innerProduct = x1[0]*x2[0] + x1[1]*x2[1]
    return innerProduct/(euc(x1, (0,0)) * euc(x2, (0,0)))
```

### k개를 찾아야함 similarity를 구해야함

```python
K = 9  # 보통 sqrt(N) # N = 짝수 K = 홀수 # C = 짝수 -> K = 홀수
candidates = sorted(list(enumerate(similarity)),
                       key=lambda x:x[1], reverse=True)[:K]
# [(i, s) for i, s in zip(range(N), similarity)]

candidateKeys = [c[0] for c in candidates]
candidateClass = list(0 for _ in range(2))

for i in candidateKeys:
    candidateClass[C[i]] += 1

    for i in range(len(C)):
        plt.scatter(X[i], Y[i],
                        facecolor = "k" if i in candidateKeys else "none",
                        edgecolors = colorMap[C[i]],
                        linewidths = .5)
        # 유사도 쓰기
        if i in candidateKeys:
            plt.text(X[i], Y[i], candidates[candidateKeys.index(i)][1])

# 점색깔        
plt.scatter(sample[0], sample[1],
       color = "r" if candidateClass[0] > candidateClass[1] else "g",
       edgecolors="b")
```
