```python
from sklearn import datasets
data = datasets.load_boston()
```

```python
len(data.data), data.feature_names # 13 x 506
```

```python
X = data.data[:,0]
Y = data.target
```

```python
X.ndim, Y.ndim, X.shape, Y.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y)
plt.show()
```

```python
import numpy as np
X = np.c_[np.ones(len(X)), X]
```

```python
X[0]
```

```python
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
```

```python
theta # bias, 기울기
```

```python
plt.scatter(X[:, 1], Y)
plt.plot(X[:, 1], X.dot(theta), "r-")
plt.show()
```

```python
X = np.c[X, np.power(X[:, 1], 2)]
```

```python
X = np.c_[X, np.power(X[:, 1], 3), np.power(X[:, 1], 4),
          np.power(X[:, 1], 5), np.power(X[:, 1], 6),
          np.power(X[:, 1], 7), np.power(X[:, 1], 8)]
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
```

```python
plt.scatter(X[:, 1], Y)
plt.scatter(X[:,1], X.dot(theta), edgecolors="r")
plt.show()
```
